```python
from pathlib import Path

base_dir = Path(".").resolve()
while base_dir != base_dir.parent and not (base_dir / "14_30").exists():
    base_dir = base_dir.parent
if not (base_dir / "14_30").exists():
    base_dir = Path(r"e:\TDTU\NCKH\Data")  # fallback đường dẫn tuyệt đối

folders = ["14_30", "15_00","15_30","16_00","16_30","17_00","17_30"]
deleted_count = 0

for folder_name in folders:
    folder = base_dir / folder_name
    if not folder.exists():
        print(f"Bỏ qua (không tồn tại): {folder}")
        continue
    for json_file in folder.rglob("*.json"):
        try:
            json_file.unlink()
            deleted_count += 1
            if deleted_count <= 5 or deleted_count % 500 == 0:
                print(f"Đã xóa: {json_file.relative_to(base_dir)}")
        except OSError as e:
            print(f"Lỗi khi xóa {json_file}: {e}")

print(f"\nTổng cộng đã xóa {deleted_count} file .json trong {folders}.")
```
